```python
# 1. Gerekli kütüphaneler
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Embedding, Conv1D, GlobalMaxPooling1D, Dense, Dropout,
                                     BatchNormalization, SpatialDropout1D, GlobalAveragePooling1D,
                                     Concatenate, Input, MaxPooling1D)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.regularizers import l2
from tensorflow.keras.models import Model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
import re
import string
import pickle
import os
from sklearn.metrics import classification_report, confusion_matrix, f1_score, precision_score, recall_score
import seaborn as sns
from tensorflow.keras.layers import LeakyReLU


# 2. Parametreler - ULTRA İYİLEŞTİRMELER
file_path = '/kaggle/input/turkish-comments-dataset/comments.csv'
model_save_path = '/kaggle/working/CNN_balanced_model_v6.keras'
checkpoint_path = '/kaggle/working/best_balanced_checkpoint_cnn_v6.keras'

max_words = 25000  # Daha da artırıldı - çok zengin vocabulary
max_len = 80       # Daha da azaltıldı - overfitting'i ciddi azaltır
embedding_dim = 200 # Ciddi artırım - daha güçlü representation
batch_size = 128   # Büyük batch - daha stabil gradientler
epochs = 25        # Daha uzun eğitim
validation_split = 0.15  # Daha az validation - daha fazla eğitim
test_size = 0.12   # Çok az test - maksimum eğitim verisi

# 3. ULTRA Geliştirilmiş metin temizleme + Data Augmentation
def clean_text(text):
    if pd.isna(text):
        return ""
    text = str(text).lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'@\w+|#\w+', '', text)
    # Türkçe karakterleri koru, sayıları da koru, noktalama işaretlerini kısmen koru
    text = re.sub(r'[^a-zA-ZçğıöşüÇĞIİÖŞÜ0-9\s\.\!\?]', ' ', text)
    # Çoklu boşlukları tek boşluk yap
    text = re.sub(r'\s+', ' ', text)
    text = text.strip()
    return text

# Basit data augmentation fonksiyonu
def augment_text(text, prob=0.1):
    """Basit kelime düzeyinde augmentation"""
    words = text.split()
    if len(words) < 3:
        return text
    
    # %10 ihtimalle kelimeleri karıştır
    if np.random.random() < prob and len(words) > 2:
        # Rastgele 2 kelimenin yerini değiştir
        idx1, idx2 = np.random.choice(len(words), 2, replace=False)
        words[idx1], words[idx2] = words[idx2], words[idx1]
    
    return ' '.join(words)

# 4. Veri yükleme ve ULTRA temizleme + Augmentation
df = pd.read_csv(file_path, encoding='utf-8')
df['text_clean'] = df['text'].apply(clean_text)
# Daha gevşek filtreleme - daha fazla veri
df = df[df['text_clean'].str.len() > 2].reset_index(drop=True)

# Data augmentation uygula (sadece azınlık sınıflar için)
original_texts = df['text_clean'].values
original_labels = df['label'].values

# Sınıf dağılımını kontrol et
label_counts = df['label'].value_counts()
print(f"Orijinal sınıf dağılımı:\n{label_counts}")

# Negative sınıfı (azınlık) için augmentation
negative_df = df[df['label'] == 'Negative'].copy()
if len(negative_df) > 0:
    # Negative örnekleri 2 kat artır
    aug_negative = negative_df.copy()
    aug_negative['text_clean'] = aug_negative['text_clean'].apply(lambda x: augment_text(x, 0.15))
    df = pd.concat([df, aug_negative], ignore_index=True)

texts = df['text_clean'].values
labels = df['label'].values
le = LabelEncoder()
labels_encoded = le.fit_transform(labels)

print(f"Augmentation sonrası sınıf dağılımı:")
print(pd.Series(labels).value_counts())

# Class weight hesaplama
class_weights = compute_class_weight('balanced', classes=np.unique(labels_encoded), y=labels_encoded)
class_weights_dict = dict(enumerate(class_weights))
print(f"Class weights: {class_weights_dict}")

# Tokenizer ayarları iyileştirildi
tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>", lower=True, 
                     filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n')
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post', truncating='post')

X_train, X_test, y_train, y_test = train_test_split(
    padded_sequences, labels_encoded, test_size=test_size, random_state=42, stratify=labels_encoded
)

# 5. ULTRA MODEL İYİLEŞTİRMESİ - Advanced Multi-scale CNN + Attention
def create_ultra_cnn_model():
    input_layer = Input(shape=(max_len,))
    
    # Güçlü embedding layer
    embedding = Embedding(max_words, embedding_dim, input_length=max_len, 
                         embeddings_regularizer=l2(0.00001))(input_layer)
    embedding = SpatialDropout1D(0.15)(embedding)  # Daha az dropout
    
    # ULTRA Multi-scale konvolüsyon - 6 farklı kernel boyutu
    conv_layers = []
    for kernel_size in [2, 3, 4, 5, 6, 7]:
        conv = Conv1D(128, kernel_size, activation='relu', 
                     kernel_regularizer=l2(0.00005), padding='same')(embedding)
        conv = BatchNormalization()(conv)
        conv = Dropout(0.2)(conv)
        
        # İkinci konvolüsyon katmanı (daha derin)
        conv = Conv1D(96, kernel_size, activation='relu', 
                     kernel_regularizer=l2(0.00005), padding='same')(conv)
        conv = BatchNormalization()(conv)
        conv = MaxPooling1D(2)(conv)
        conv = Dropout(0.2)(conv)
        
        conv_layers.append(conv)
    
    # Attention benzeri mekanizma için global pooling çeşitleri
    pooled_layers = []
    for conv in conv_layers:
        # Her konvolüsyon için hem max hem average pooling
        max_pool = GlobalMaxPooling1D()(conv)
        avg_pool = GlobalAveragePooling1D()(conv)
        pooled_layers.extend([max_pool, avg_pool])
    
    # Tüm feature'ları birleştir
    concatenated = Concatenate()(pooled_layers)
    
    # Çok güçlü dense katmanlar
    dense1 = Dense(512, kernel_regularizer=l2(0.00005))(concatenated)
    dense1 = BatchNormalization()(dense1)
    dense1 = LeakyReLU(alpha=0.1)(dense1)
    dense1 = Dropout(0.4)(dense1)
    
    dense2 = Dense(256, kernel_regularizer=l2(0.00005))(dense1)
    dense2 = BatchNormalization()(dense2)
    dense2 = LeakyReLU(alpha=0.1)(dense2)
    dense2 = Dropout(0.35)(dense2)
    
    dense3 = Dense(128, kernel_regularizer=l2(0.00005))(dense2)
    dense3 = BatchNormalization()(dense3)
    dense3 = LeakyReLU(alpha=0.1)(dense3)
    dense3 = Dropout(0.3)(dense3)
    
    dense4 = Dense(64, kernel_regularizer=l2(0.00005))(dense3)
    dense4 = BatchNormalization()(dense4)
    dense4 = LeakyReLU(alpha=0.1)(dense4)
    dense4 = Dropout(0.25)(dense4)
    
    # Output layer
    output = Dense(len(le.classes_), activation='softmax')(dense4)
    
    model = Model(inputs=input_layer, outputs=output)
    return model

# Model oluştur
model = create_ultra_cnn_model()

# ULTRA OPTIMIZER İYİLEŞTİRMESİ
initial_learning_rate = 0.003  # Daha yüksek başlangıç
optimizer = tf.keras.optimizers.Adam(
    learning_rate=initial_learning_rate,
    beta_1=0.9,
    beta_2=0.999,
    epsilon=1e-8,
    clipnorm=0.5  # Daha agresif gradient clipping
)

model.compile(
    loss='sparse_categorical_crossentropy', 
    optimizer=optimizer, 
    metrics=['accuracy']
)

model.summary()

# 6. ULTRA İyileştirilmiş Callback'ler
callbacks = [
    EarlyStopping(
        monitor='val_loss', 
        patience=7,  # Daha fazla patience
        restore_best_weights=True, 
        verbose=1,
        min_delta=0.0005  # Daha hassas threshold
    ),
    ReduceLROnPlateau(
        monitor='val_loss', 
        factor=0.4,  # Daha yumuşak azaltma
        patience=4,  # Daha sabırlı
        min_lr=1e-9, 
        verbose=1, 
        cooldown=3,  # Daha uzun cooldown
        min_delta=0.0005
    ),
    ModelCheckpoint(
        checkpoint_path, 
        monitor='val_accuracy',
        save_best_only=True, 
        verbose=1, 
        mode='max',
        save_weights_only=False
    )
]

# 7. KRİTİK EĞİTİM İYİLEŞTİRMESİ
# Data augmentation için shuffle=True ve daha büyük batch size
history = model.fit(
    X_train, y_train,
    epochs=epochs,
    batch_size=batch_size,
    validation_split=validation_split,
    callbacks=callbacks,
    verbose=1,
    shuffle=True,
    class_weight=class_weights_dict,
    # Validation frequency ayarı
    validation_freq=1
)

# 8. Tokenizer & LabelEncoder kaydet
with open('/kaggle/working/tokenizer_balanced_v6.pickle', 'wb') as handle:
    pickle.dump(tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)
with open('/kaggle/working/labelencoder_balanced_v6.pickle', 'wb') as handle:
    pickle.dump(le, handle, protocol=pickle.HIGHEST_PROTOCOL)

# 9. Eğitim Grafikleri
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

axes[0,0].plot(history.history['accuracy'], label='Eğitim Doğruluğu', marker='o')
axes[0,0].plot(history.history['val_accuracy'], label='Doğrulama Doğruluğu', marker='s')
axes[0,0].set_title('Model Doğruluğu')
axes[0,0].legend()
axes[0,0].grid(True)

axes[0,1].plot(history.history['loss'], label='Eğitim Kaybı', marker='o')
axes[0,1].plot(history.history['val_loss'], label='Doğrulama Kaybı', marker='s')
axes[0,1].set_title('Model Kaybı')
axes[0,1].legend()
axes[0,1].grid(True)

if 'lr' in history.history:
    axes[1,0].plot(history.history['lr'], marker='o', color='red')
    axes[1,0].set_title('Learning Rate')
    axes[1,0].set_yscale('log')
    axes[1,0].grid(True)
else:
    axes[1,0].text(0.5, 0.5, f'Toplam Epoch: {len(history.history["loss"])}',
                   transform=axes[1,0].transAxes, ha='center', va='center', fontsize=16)

train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
gap = [t - v for t, v in zip(train_acc, val_acc)]
axes[1,1].plot(gap, marker='o', color='orange')
axes[1,1].axhline(0, linestyle='--', color='black', alpha=0.5)
axes[1,1].set_title('Overfitting Gap')
axes[1,1].grid(True)

plt.tight_layout()
plt.show()

# 10. KRİTİK TEST DEĞERLENDİRMESİ
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
test_predictions = model.predict(X_test, verbose=0)
test_pred_classes = np.argmax(test_predictions, axis=1)

print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print("\nSınıf Bazında Performans:")
print(classification_report(y_test, test_pred_classes, target_names=le.classes_))

f1_macro = f1_score(y_test, test_pred_classes, average='macro')
f1_weighted = f1_score(y_test, test_pred_classes, average='weighted')
f1_micro = f1_score(y_test, test_pred_classes, average='micro')
precision_macro = precision_score(y_test, test_pred_classes, average='macro')
recall_macro = recall_score(y_test, test_pred_classes, average='macro')

print(f"Macro F1-score: {f1_macro:.4f}")
print(f"Weighted F1-score: {f1_weighted:.4f}")
print(f"Micro F1-score: {f1_micro:.4f}")
print(f"Macro Precision: {precision_macro:.4f}")
print(f"Macro Recall: {recall_macro:.4f}")

cm = confusion_matrix(y_test, test_pred_classes)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Tahmin')
plt.ylabel('Gerçek')
plt.title('Confusion Matrix - ULTRA İyileştirilmiş Model')
plt.show()

# 11. Eğitim Özeti
train_acc_final = history.history['accuracy'][-1]
val_acc_final = history.history['val_accuracy'][-1]
overfitting_gap_final = train_acc_final - val_acc_final
total_params = model.count_params()

print(f"\nFinal Eğitim Accuracy: {train_acc_final:.4f}")
print(f"Final Validation Accuracy: {val_acc_final:.4f}")
print(f"Overfitting Gap: {overfitting_gap_final:.4f}")
print(f"Toplam Parametre: {total_params:,}")

train_test_gap = train_acc_final - test_accuracy
print(f"Eğitim - Test Gap: {train_test_gap:.4f}")

# 12. Model Kaydet
model.save(model_save_path)
model_size = os.path.getsize(model_save_path) / (1024 * 1024)
print(f"Model {model_save_path} konumuna kaydedildi. ({model_size:.2f} MB)")

# 13. İyileştirilmiş Metrik Grafiği
metrics = {
    "Test\nAccuracy": test_accuracy,
    "F1-score\n(macro)": f1_macro,
    "F1-score\n(weighted)": f1_weighted,
    "Precision\n(macro)": precision_macro,
    "Recall\n(macro)": recall_macro
}

plt.figure(figsize=(12,6))
bars = plt.bar(metrics.keys(), metrics.values(), 
               color=['#2E8B57', '#FF6B35', '#4682B4', '#9932CC', '#FF1493'])
plt.ylim(0, 1.0)
plt.title("Test Seti Metrikleri (ULTRA Multi-Scale CNN)", fontsize=14, fontweight='bold')
plt.ylabel("Skor")
plt.grid(axis='y', alpha=0.3)
for bar, value in zip(bars, metrics.values()):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f'{value:.3f}', ha='center', va='bottom', fontweight='bold')
plt.tight_layout()
plt.show()

# 14. Gelişmiş Performans Özeti
print("="*70)
print("MODEL PERFORMANS ÖZETİ (ULTRA Multi-Scale CNN + Data Aug)")
print("="*70)
print(f"🎯 Test Accuracy: {test_accuracy:.1%}")
print(f"🎯 F1-Score (Macro): {f1_macro:.1%}")
print(f"🎯 F1-Score (Weighted): {f1_weighted:.1%}")
print(f"🎯 Overfitting Gap: {overfitting_gap_final:.1%}")
print(f"📊 Model Parametreleri: {total_params:,}")
print(f"💾 Model Boyutu: {model_size:.1f} MB")
print(f"⏱️  Eğitim Epoch: {len(history.history['loss'])}")
print(f"⚖️  Class Weight + Data Augmentation Kullanıldı")
print(f"🔧 Architecture: 6-Scale CNN + Deep Dense + Attention-like")
print(f"📊 Vocabulary Size: {max_words:,}")
print(f"📏 Sequence Length: {max_len}")

performance_score = (test_accuracy + f1_weighted + f1_macro) / 3
if performance_score > 0.92 and overfitting_gap_final < 0.03:
    print("\n🏆 MÜKEMMEL MODEL! En yüksek seviye performans!")
elif performance_score > 0.88 and overfitting_gap_final < 0.04:
    print("\n✅ ÇOK İYİ MODEL! Profesyonel seviye performans")
elif performance_score > 0.85:
    print("\n👍 İYİ MODEL! Güçlü performans")
else:
    print("\n⚠️  Model daha fazla iyileştirme gerektirebilir")
    
print(f"\n📈 ULTRA Performans Skoru: {performance_score:.1%}")
print(f"🎯 Macro F1 Hedefi: >91% (Mevcut: {f1_macro:.1%})")
print(f"🎯 Overfitting Hedefi: <3% (Mevcut: {overfitting_gap_final:.1%})")
```
